```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 

data = pd.read_csv('iris.data', header=None)
```

```python
X = data.iloc[:,:4].to_numpy()
labels = data.iloc[:,4].to_numpy()
```

```python
le = preprocessing.LabelEncoder()
Y = le.fit_transform(labels)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
print((y_test != y_pred).sum()) #количество наблюдений, который были неправильно определены
```

```python
accuracy = gnb.score(X_test, y_test)
print("Точность:", accuracy)
```

```python
random_state = 5

test_sizes = np.arange(0.05, 1.0, 0.05)

count_misclassified = []
accuracies = []

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=random_state)

    gnb = GaussianNB()
    y_pred = gnb.fit(X_train, y_train).predict(X_test)

    misclassified = (y_test != y_pred).sum()
    accuracy = gnb.score(X_test, y_test)

    count_misclassified.append(misclassified)
    accuracies.append(accuracy)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(test_sizes, count_misclassified, marker='o', color='g')
plt.title('Количество неправильно классифицированных наблюдений')
plt.xlabel('Размер')
plt.ylabel('Неправильно классифицированные наблюдения')

plt.subplot(1, 2, 2)
plt.plot(test_sizes, accuracies, marker='o', color='r')
plt.title('Точность классификации')
plt.xlabel('Размер')
plt.ylabel('Точность')

plt.tight_layout()
plt.show()
```
