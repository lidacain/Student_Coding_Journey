```python
from scipy.stats import shapiro
import numpy as np

df = pd.read_excel('Qz.xls', sheet_name='year')

# save the first column as a list
names = df.iloc[:, 0].values.tolist()

df.replace('...', np.nan, inplace=True)

# convert the columns to numeric type
df = df.apply(pd.to_numeric, errors='coerce')

# replace NaN values with the mean of the corresponding column
df.fillna(df.mean(), inplace=True)

# change the first column to the list of names
df.iloc[:, 0] = names

df

```

```python
stat, p_val = shapiro(df['Көкшетау'])

print(f"Shapiro-Wilk test statistic: {stat:.3f}")
print(f"p-value: {p_val:.3f}")
if p_val > 0.05:
    print("The column is likely normally distributed.")
else:
    print("The column is likely not normally distributed.")
```

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_excel('Qz.xls', sheet_name='4months')
df = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

group1 = df['Алматы']
group2 = df['Астана']
t, p = stats.ttest_ind(group1, group2)

print("Independent t-test:")
print(f"t-statistic: {t}")
print(f"p-value: {p}")

df = pd.read_excel('Qz.xls', sheet_name='prodPercent')
df = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

time1 = df['бір жыл ішінде']
time2 = df['жыл басынан\nбері кезеңге']
t, p = stats.ttest_rel(time1, time2)

print("Dependent t-test:")
print(f"t-statistic: {t}")
print(f"p-value: {p}")
```
